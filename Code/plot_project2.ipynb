```python
def format_plot(x_title, y_title, title):
    plt.style.use('ggplot')
    plt.rcParams['axes.facecolor']='white'
    plt.rc('grid', color='#ECECEC', linestyle='solid')
    plt.xticks(color='#626F78', rotation='0', fontweight='bold', fontsize='12')
    plt.yticks(color='#626F78', fontweight='bold', fontsize='14', horizontalalignment='right')
    plt.xlabel(x_title, labelpad = 20, fontweight='bold', color = '#626F78', fontsize='17', 
               horizontalalignment='center')
    plt.ylabel(y_title, labelpad = 20, fontweight='bold', color = '#626F78', fontsize='17', 
               horizontalalignment='center')
    plt.title(title, color = "#626F78", loc='center', horizontalalignment='center', fontsize = '20', y= 1.05, 
              fontweight='bold',verticalalignment='center')
    plt.xticks(color='#626F78', rotation='0', fontweight='bold', fontsize='14')
```

```python
by_tbin = dfm.groupby('temp_bin')
```

```python
by_tbin.get_group('Cold')
```

```python
def stars_ptg(gr):
    return gr.stars.value_counts(normalize=True)

starn = by_tbin.apply(stars_ptg).sort_values(by='stars', ascending = True, axis=1)
```

```python
starn
```

```python
#starn.plot(kind='bar',stacked=True,cmap='tab10', figsize=(10,5))
starn.plot(kind='bar',stacked=True,colors=['darkgray',  'bisque', 'darkseagreen', 'palevioletred', (0.2, 0.4, 0.6, 0.6)],
           figsize=(10,5))
format_plot('Weather','Ratio','Restaurant Reviews by Temperature')
plt.legend(loc='center left', bbox_to_anchor= (1.0, 0.5), ncol=1, borderaxespad=0, frameon=False,fontsize = 15)
#plt.legend(('1','2','3','4','5'),loc='upper right')#, title='Review Stars')
#plt.title('Restaurant Reviews by Temperature', fontsize=25)
#plt.xlabel('Weather', fontsize=18)
#plt.ylabel('Percentage', fontsize=18)
plt.savefig('reviews_temp.png',transparent=True, bbox_inches='tight', pad_inches=0)
```
