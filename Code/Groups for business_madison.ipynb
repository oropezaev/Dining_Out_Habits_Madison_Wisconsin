```python
cuisine_dict = {
    'asian' : 'asian',
    'chinese' : 'asian',
    'sushi' : 'asian',
    'japanese' : 'asian',
    'korean' : 'asian',
    'thai' : 'asian',
    'taiwan' : 'asian',
    'mongolian' : 'asian',
    'vietnamese' : 'asian',
    'indonesian' : 'asian',
    'middle eastern' : 'asian',
    'persian' : 'asian',
    'latin' : 'latin american',
    'brazilian' : 'latin american',
    'mexican' : 'latin american',
    'caribbean' : 'latin american',
    'peruvian' : 'latin american',
    'indian' : 'asian',
    'nepalese' : 'asian',
    'afghan' : 'asian',
    'pizza' : 'american',
    'ice cream' : 'american',
    'sandwich' : 'american',
    'deli' : 'american',
    'steak' : 'american',
    'burgers' : 'american',
    'seafood' : 'american',
    'fast food' : 'american',
    'mediterranean' : 'european',
    'greek' : 'european',
    'french' : 'european',
    'german' : 'european',
    'belgian' : 'european',
    'italian' : 'european',
    'coffee' : 'american',
    'bakeries' : 'american',
    'breakfast' : 'american',
    'cafe' : 'american',
    'cajun' : 'european',
    'southern' : 'american',
    'soul' : 'american',
    'chicken wings' : 'american',
    'comfort' : 'american',
    'russian' : 'other',
    'moroccan' : 'other',
    'african' : 'other',
    'american' : 'american',
}


soup_dict = {
    'soup' : 'soup',
    'ramen' : 'soup'
}

comfort_dict = {
    'southern' : 'comfort',
    'soul' : 'comfort',
    'chicken wings' : 'comfort',
    'comfort' : 'comfort',
}

zip_dict = {
    53175 : 'unknown',
    53518 : 'southwest',
    53562 : 'southwest',
    53593 : 'southwest',
    53702 : 'northeast',
    53703 : 'between lakes',
    53704 : 'northeast',
    53705 : 'southwest',
    53706 : 'southwest',
    53711 : 'southwest',
    53713 : 'southwest',
    53714 : 'northeast',
    53715 : 'southwest',
    53716 : 'northeast',
    53717 : 'southwest',
    53718 : 'northeast',
    53719 : 'southwest',
    53726 : 'southwest',
    53774 : 'between lakes'
}
```

```python
def cuisine(a):
    out = 'other'
    for key in cuisine_dict.keys():
        if key in a.lower():
            out = cuisine_dict[key]
            break
    return out

def soup_fun(a):
    out = 'other'
    for key in soup_dict.keys():
        if key in a.lower():
            out = soup_dict[key]
            break
    return out

def comfort_fun(a):
    out = 'other'
    for key in comfort_dict.keys():
        if key in a.lower():
            out = comfort_dict[key]
            break
    return out

def amer(a):
    out = 'other'
    if a.lower() == 'american':
        out = 'american'
    return out

def zip_area(a):
    out = 'other'
    for key in zip_dict.keys():
        if a == key:
            out = zip_dict[key]
            break
    return out
```

```python
grouped = bs.categories.apply(cuisine)
bs['cuisine_grp'] = grouped

american = bs.cuisine_grp.apply(amer)
bs['american'] = american

soup = bs.categories.apply(soup_fun)
bs['soup'] = soup

comfort = bs.categories.apply(comfort_fun)
bs['comfort'] = comfort

zip_areas = bs.postal_code.apply(zip_area)
bs['zip_area'] = zip_areas
```
